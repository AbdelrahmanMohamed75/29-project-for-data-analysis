### Step 5. Assign the columns as below:

The attributes are (donated by Riccardo Leardi, riclea '@' anchem.unige.it):  
1) alcohol  
2) malic_acid  
3) alcalinity_of_ash  
4) magnesium  
5) flavanoids

7) hue 

```python
wine['alcohol']=wine['14.23']
wine['malic_acid']=wine['1.71']
wine['alcalinity_of_ash']=wine['15.6']
wine['magnesium']=wine['127']
wine['flavanoids']=wine['3.06']
wine['proanthocyanins']=wine['2.29']
wine['hue']=wine['1.04']
wine.drop(columns=['14.23','1.71','15.6','127','3.06','2.29','1.04'],inplace=True)
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine.loc[:2,'alcohol']=float('nan')
wine
```
