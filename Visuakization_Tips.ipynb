```python
tips['total_bill'].plot(kind='hist')
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips['total_bill'],tips['tip'])
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('relationship between total_bill and tip')

plt.grid(True)
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.bar(tips['day'],tips['total_bill'])
plt.xlabel('days')
plt.ylabel('total_bill')
plt.title('relationship between days and total_bill value')
plt.grid(True)
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
import seaborn as sns
sns.scatterplot(data=tips,x='day',y='tip',hue='sex')
plt.xlabel('Tip ($)')
plt.ylabel('Day')
plt.title('Tip vs. Day Differed by Sex')
plt.legend(title='sex')
plt.grid(True)
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(data=tips,x='total_bill',y='day',hue='time')
plt.xlabel('total_bill')
plt.ylabel('day')
plt.legend(title='time')
plt.title(' presenting the total_bill per day differetiation the time (Dinner or Lunch)')
plt.grid(True)
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
dinner_tips = tips[tips['time'] =='Dinner']['tip']
lunch_tips = tips[tips['time'] == 'Lunch']['tip']

# Create subplots
plt.figure(figsize=(12, 6))

# Plot histogram for Dinner
plt.subplot(1, 2, 1)
plt.hist(dinner_tips, bins=10, color='skyblue', edgecolor='black')
plt.title('Tip Distribution for Dinner')
plt.xlabel('Tip ($)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(lunch_tips, bins=10, color='salmon', edgecolor='black')
plt.title('Tip Distribution for Lunch')
plt.xlabel('Tip ($)')
plt.ylabel('Frequency')
```
